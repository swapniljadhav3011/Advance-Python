List vs Array multiplication of numbers


```python
b = [3, 4, 5, 6, 7]
type(b)
```

```python
b*3
```

```python
b*2
```

```python
# You need to apply for loop to multiply all numbers by 3
c = []
for i in b:
    c.append(i*3)
```

```python
c
```

```python
d = np.array(b)
d
```

```python
d*3
```

```python
d*2.5
```

```python
e = 2.1*d + 1.5
```

```python
e
```

## Get product of all elements inside list


```python
t = [2, 3, 4, 5, 6]
type(t)
```

```python
p = 1
for i in t:
    p = p*i
```
